##Matrix Multiplication

Two main ways of performing multiplication in neural networks and deep learning:
1. element-wise multiplication
2. matrix multiplication

there are two more rules that performing matrix multiplication needs to satisfy:
1. the **inner dimensions** must match:
* (3, 2) @ (3, 2) wont work
* (2, 3) @ (3, 2) will work
* (3, 2) @ (2, 3) will work

2. the resulting matrix has the shape of **outer dimentions**:
* (2, 3) @ (3, 2) => (2, 2)
* (3, 2) @ (2, 3) => (3, 3)

```python
#matrix_multiplication torch.matmul(tensor_1, tensor_2)
torch.matmul(tensor, tensor)
```

```python
tensor @ tensor
```

## Reshaping, stacking, squeesing, unsqueezzing tensors

* Reshaping  - Reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - Combine multiplle tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - Removes all '1' dimensions from a tensor
Unsqueeze - adds a 'a' dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1., 10.)
x , x.shape
```

```python
#Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Changing the view
z = x.view(3,3)
z, z.shape
```

Changing **z** changes **x** (because a view of a tensor shares the same memory as the original input)

```python
z[:, 0] = 5
z, x
```

```python
#Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x], dim=1)
x_stacked
```

```python
#torch.squeeze - Removes all single dimensions from a target tensor
x = torch.zeros(2,1,2,1,2)
print(x.shape)

y = torch.squeeze(x)
print(y.shape)

z = torch.squeeze(x, 0)
print(z.shape)

k = torch.squeeze(x, 1)
print(k.shape)
```

```python
print(f"the default tensor : {x_reshaped}")
print(f"the shape of the tensor : {x_reshaped.shape}")
```

```python
print(f"the tensor after being squeezed : {x_reshaped.squeeze()}")
print(f"the size of the squeezed tensor : {x_reshaped.squeeze().shape}")
x_squeezed = x_reshaped.squeeze()
```

###Squeeze removes Extra dimensions
###UnSqueeze = torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim (dimension)

```python
print(f"previous target : {x_squeezed}")
print(f"previous shape : {x_squeezed.shape}")

#Adding an extra dimension with torch.unsqueeze()
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew Tensor: {x_unsqueezed}")
print(f"New Shape : {x_unsqueezed.shape}")
```

#torch.permute - rearranges the dimensions of a target tensor in a specified order

```python
x_original = torch.rand(size=(224, 224, 3)) #[height, width, colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) #shifts axis 0=>1, 1=>2 , 2=>0

print(f"Previous shape: {x_original.shape}")
print(f"New shape : {x_permuted.shape}") #[colour_channels, height, width]
```

## Pytorch tensors & NumPy

NumPy is a popular scientific Python numerical computing library.

* Data in NumPy, want in Pytorch tensor => 'torch.from_numpy(ndarrray)'
*PyTorch tensor => NumPy => 'torch.Tensor.numpy()'

```python
#Numpy array to tensor
import torch
import numpy as np

array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array) #warning when converting from numpy to pytorch, Pytorch reflects numpys default datatype of float64
array, tensor
```

## Reproducbility (trying to take random out of random)

in short how a neural network learns:
start with random numbers => tensor operations => update random numbers to try and make them better representations of the data => again => again => again...

to reduce the randimness in neural networks and pytorch comes the concept of a **random seed**

essiantly what the random seed does is "flavour" the randomness.


```python
import torch
#create 2 random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)

```

```python
#Lets make some random but reproducible tensors

#set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED) #use everytime you create a random torch.tensor
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
